```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],axis=1, inplace= True)
```

```python
covid_df.head()
```

```python
covid_df["Date"]= pd.to_datetime(covid_df['Date'], format= '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
covid_df['Active_cases']= covid_df['Confirmed']- (covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.head(10)
```

```python
statewise= pd.pivot_table(covid_df, values= ['Confirmed', 'Deaths', 'Cured'], index= 'State/UnionTerritory', aggfunc=max)
statewise['recovery_rate']= statewise['Cured'] *100 / statewise['Confirmed']
statewise['mortality_rate']= statewise['Deaths'] *100 / statewise['Confirmed']
statewise= statewise.sort_values(by= 'Confirmed', ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

### Top 10 active cases

```python
top_10_active_cases= covid_df.groupby(by= 'State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'], ascending=False).reset_index()
```

```python
top_10_active_cases.head(10)
```

```python
plt.figure(figsize= (12,5))
sns.barplot(data= top_10_active_cases.iloc[:10],x= 'State/UnionTerritory', y= 'Active_cases',palette = 'flare', linewidth =2, edgecolor= 'red')
plt.xlabel('States')
plt.ylabel('Active_cases')
plt.title('Top 10 states with most active cases in India')
plt.show()
```

```python
top_10_death_cases= covid_df.groupby(by= 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by= ['Deaths'], ascending = False).reset_index()
```

```python
plt.figure(figsize=(14,8))
sns.barplot(data=top_10_death_cases.iloc[:10], x='State/UnionTerritory', y='Deaths', palette='Spectral', linewidth=2)
plt.xlabel('States')
plt.ylabel('Deaths of the covid_person')
plt.title('Top 10 states with most death cases in India')
plt.show()
```

### male vs female vaccination

```python
male= vaccine_df['Male(Individuals Vaccinated)'].sum()
female= vaccine_df['Female(Individuals Vaccinated)'].sum()
plt.pie([male,female], labels= ['Male','Female'], autopct='%.2f%%')
plt.title('Male vs Female Vaccination')
plt.show()
```

```python
vaccine_df['State'].value_counts()
```

```python
vaccine= vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated': 'Total'}, inplace= True)
vaccine.head()
```

### most vaccinated State

```python
max_vacc= vaccine.groupby('State')['Total'].sum().sort_values(ascending= False).reset_index()
max_vacc[:5]
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data= max_vacc.iloc[:10], y= 'Total', x= 'State', palette = 'flare', linewidth =2, edgecolor='black')
plt.title('Top Vaccinated State In India')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Vaccination')
plt.show()
```

### least vaccinated state

```python
min_vacc= vaccine.groupby('State')['Total'].sum().sort_values().reset_index()
min_vacc[:5]
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data= min_vacc.iloc[:10], y='Total', x='State', palette='cool', linewidth=2, edgecolor='red')
plt.title('Least Vaccinated State in India')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Vaccination')
plt.show()
```
